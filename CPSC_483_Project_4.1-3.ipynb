### 2) Fit and score binary classifiers for dataset 1 using logistic regression, Gaussian Naive Bayes and K-Nearest Neighbors for dataset1.

How accurate are the model's predictions?

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

#splitting x from y for dataset1
df1_x = df1.iloc[:,:2]
#print(df1_x)

df1_y = df1.iloc[: , 2:3]
#print(df1_y)


#Logistic Regression
df1_LR_classifier = LogisticRegression()
df1_LR_classifier.fit(df1_x, np.ravel(df1_y))
df1_LR_score = df1_LR_classifier.score(df1_x, df1_y)
print("LR Score for Dataset1: ", df1_LR_score)

#Gaussian Naive Bayes
df1_GNB_classifier = GaussianNB()
df1_GNB_classifier.fit(df1_x, np.ravel(df1_y))
df1_GNB_score = df1_GNB_classifier.score(df1_x, df1_y)
print("GNB Score for Dataset1: ", df1_GNB_score)

#KNN
df1_KNN_classifier = KNeighborsClassifier()
df1_KNN_classifier.fit(df1_x, np.ravel(df1_y))
df1_KNN_score = df1_KNN_classifier.score(df1_x, df1_y)
print("KNN Score for Dataset1: ", df1_KNN_score)
```
